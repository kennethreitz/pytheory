# PyTheory: Music Theory for Humans

A hands-on tutorial exploring music theory with Python.

PyTheory lets you reason about tones, scales, chords, and progressions
using an intuitive, Pythonic API. Whether you're a musician who codes
or a coder who plays music, this library gives you the building blocks
to explore harmony, composition, and world music systems.

## 1. Getting Started

Everything begins with a **Tone** -- the fundamental unit of music.
A tone has a name (like `C`, `F#`, or `Bb`), an optional octave number,
and a frequency in Hz computed from equal temperament tuning (A4 = 440 Hz).

```python
from pytheory import Tone, TonedScale, Key, Chord, Fretboard, CHARTS, Interval
from pytheory import analyze_progression
from pytheory.scales import PROGRESSIONS
```

```python
# Create tones with octave numbers (scientific pitch notation)
middle_c = Tone.from_string("C4")
concert_a = Tone.from_string("A4")

print(f"Middle C:  {middle_c}  ->  {middle_c.frequency:.2f} Hz")
print(f"Concert A: {concert_a}  ->  {concert_a.frequency:.2f} Hz")
print(f"MIDI note: {middle_c.midi}")
print(f"Is natural? {middle_c.is_natural}")
```

```python
# Create tones from frequencies or MIDI numbers
from_hz = Tone.from_frequency(440.0)
from_midi = Tone.from_midi(60)

print(f"440 Hz -> {from_hz}")
print(f"MIDI 60 -> {from_midi}")
```

```python
# Explore the harmonic series -- the physics behind consonance
c3 = Tone.from_string("C3")
harmonics = c3.overtones(8)
print(f"Harmonic series of {c3} ({c3.frequency:.1f} Hz):")
for i, hz in enumerate(harmonics, 1):
    print(f"  Harmonic {i}: {hz:.1f} Hz")
```

## 2. Tone Arithmetic

Tones support arithmetic operations. Adding an integer to a tone raises it
by that many **semitones** (half steps). Subtracting two tones gives the
semitone distance between them. You can also compare tones by pitch.

```python
c4 = Tone.from_string("C4")

# Add semitones: C + 4 semitones = E (a major third)
e4 = c4 + 4
g4 = c4 + Interval.PERFECT_FIFTH
print(f"{c4} + 4 semitones  = {e4}")
print(f"{c4} + perfect 5th  = {g4}")

# Subtract to find interval distance
distance = g4 - c4
print(f"\nDistance from {c4} to {g4}: {distance} semitones")
```

```python
# Name the interval between two tones
print(f"{c4} -> {e4}: {c4.interval_to(e4)}")
print(f"{c4} -> {g4}: {c4.interval_to(g4)}")

c5 = Tone.from_string("C5")
print(f"{c4} -> {c5}: {c4.interval_to(c5)}")

# Compare tones by pitch
print(f"\n{c4} < {g4}? {c4 < g4}")
print(f"{c4} == {c4}? {c4 == c4}")
```

```python
# The circle of fifths -- the backbone of Western harmony
c = Tone.from_string("C4")
fifths = c.circle_of_fifths()
print("Circle of fifths from C:")
print(" -> ".join(str(t) for t in fifths))
```

## 3. Scales and Modes

A **scale** is a collection of tones arranged in ascending order.
The `TonedScale` class provides access to dozens of scales from a given tonic.

**Modes** are rotations of the same set of intervals. The seven modes of the
major scale each have a distinct character:

| Mode       | Character          |
|------------|--------------------|
| Ionian     | Bright, happy      |
| Dorian     | Jazzy, soulful     |
| Phrygian   | Spanish, dark      |
| Lydian     | Dreamy, floating   |
| Mixolydian | Bluesy, rock       |
| Aeolian    | Sad, natural minor |
| Locrian    | Tense, unstable    |

```python
# Build a scale from a tonic
ts = TonedScale(tonic="C4")

# See all available scale names
print("Available scales:")
for name in ts.scales:
    print(f"  {name}")
```

```python
# Get a specific scale and iterate its tones
c_major = ts["major"]
print(f"C major: {c_major.note_names}")

c_minor = ts["minor"]
print(f"C minor: {c_minor.note_names}")

# Check if a note belongs to the scale
print(f"\nIs F# in C major? {'F#' in c_major}")
print(f"Is G in C major?  {'G' in c_major}")
```

```python
from pytheory.scales import Scale

# Transpose a scale
d_major = c_major.transpose(2)
print(f"D major (C major transposed up 2): {d_major.note_names}")

# Detect a scale from a set of notes
result = Scale.detect("A", "B", "C#", "D", "E", "F#", "G#")
print(f"\nDetected scale: {result}")
```

## 4. The Key Class

A **Key** is the most convenient entry point for working with harmony.
It wraps a tonic and mode, giving you instant access to scales, diatonic
chords, key signatures, and related keys.

```python
key = Key("C", "major")

print(f"Key: {key}")
print(f"Notes: {key.note_names}")
print(f"Signature: {key.signature}")
```

```python
# Diatonic triads -- the seven chords built from the scale
print("Diatonic triads in C major:")
for i, name in enumerate(key.chords, 1):
    print(f"  {i}. {name}")
```

```python
# Seventh chords add richness and color
print("Diatonic seventh chords in C major:")
for i, name in enumerate(key.seventh_chords, 1):
    print(f"  {i}. {name}")
```

```python
# Related keys
print(f"Relative minor of C major: {key.relative}")
print(f"Parallel minor of C major: {key.parallel}")

# Key signatures for sharp and flat keys
for tonic in ["G", "D", "F", "Bb"]:
    k = Key(tonic, "major")
    sig = k.signature
    print(f"{k}: {sig['sharps']} sharps, {sig['flats']} flats -> {sig['accidentals']}")
```

## 5. Chord Analysis

Chords can be created from note names, intervals, chord symbols, or MIDI.
PyTheory can identify chord quality, measure tension and consonance,
and compute optimal voice leading between chords.

```python
# Multiple ways to create chords
c_major_chord = Chord.from_tones("C", "E", "G")
g7 = Chord.from_intervals("G", 4, 7, 10)
am = Chord.from_name("Am")

print(f"{c_major_chord}  (intervals: {c_major_chord.intervals})")
print(f"{g7}  (intervals: {g7.intervals})")
print(f"{am}  (intervals: {am.intervals})")
```

```python
# Analyze harmonic tension
# The dominant 7th chord is the most tension-filled chord in tonal music
print("Tension analysis:")
for chord in [c_major_chord, am, g7]:
    t = chord.tension
    print(f"  {chord.identify():20s} -> score={t['score']:.2f}, "
          f"tritones={t['tritones']}, dominant={t['has_dominant_function']}")
```

```python
# Consonance vs dissonance (psychoacoustic measures)
print(f"{'Chord':20s} {'Harmony':>10s} {'Dissonance':>12s}")
print("-" * 44)
for chord in [c_major_chord, am, g7]:
    print(f"{chord.identify():20s} {chord.harmony:10.4f} {chord.dissonance:12.4f}")
```

```python
# Voice leading: how individual notes move between chords
f_major = Chord.from_tones("F", "A", "C")
vl = c_major_chord.voice_leading(f_major)

print(f"Voice leading: {c_major_chord.identify()} -> {f_major.identify()}")
for src, dst, movement in vl:
    direction = "up" if movement > 0 else "down" if movement < 0 else "stays"
    print(f"  {src} -> {dst} ({movement:+d} semitones, {direction})")
```

```python
# Inversions rearrange chord voicings
print(f"Root position: {[t.full_name for t in c_major_chord.tones]}")
print(f"1st inversion: {[t.full_name for t in c_major_chord.inversion(1).tones]}")
print(f"2nd inversion: {[t.full_name for t in c_major_chord.inversion(2).tones]}")
```

## 6. Chord Progressions

Chord progressions are the harmonic backbone of songs. PyTheory supports
both **Roman numeral** analysis (classical/jazz) and the **Nashville number
system** (studio shorthand). It also ships with common progressions built in.

```python
key = Key("G", "major")

# Build a progression from Roman numerals
prog = key.progression("I", "V", "vi", "IV")
print("I - V - vi - IV in G major (the 'four chord song'):")
for chord in prog:
    print(f"  {chord.identify()}")
```

```python
# Nashville number system -- same thing, Arabic numerals
nashville = key.nashville(1, 5, 6, 4)
print("Nashville 1-5-6-4 in G major:")
for chord in nashville:
    print(f"  {chord.identify()}")
```

```python
# Browse the built-in progression library
print("Built-in progressions:")
for name, numerals in PROGRESSIONS.items():
    print(f"  {name:25s} -> {' '.join(numerals)}")
```

```python
# Analyze an existing chord progression
chords = [Chord.from_name("C"), Chord.from_name("Am"),
          Chord.from_name("F"),  Chord.from_name("G")]
numerals = analyze_progression(chords, key="C")
print("Progression analysis in C:")
for chord, numeral in zip(chords, numerals):
    print(f"  {chord.identify():15s} -> {numeral}")
```

## 7. World Music Systems

Music theory extends far beyond Western harmony. PyTheory includes scale
systems from several traditions:

- **Indian** (raga/thaat) -- 10 parent scales covering all of Hindustani music
- **Arabic** (maqam) -- modal systems with characteristic augmented seconds
- **Japanese** -- pentatonic scales used in koto, shamisen, and folk music
- **Blues** -- the scales that built American popular music
- **Gamelan** -- Javanese/Balinese tuning systems (12-TET approximations)

```python
from pytheory import SYSTEMS

# Indian thaat system
indian = TonedScale(tonic="C4", system=SYSTEMS["indian"])
print("Indian thaats from C:")
for name in indian.scales:
    scale = indian[name]
    print(f"  {name:12s} -> {scale.note_names}")
```

```python
# Arabic maqam -- the Hijaz scale has a distinctive augmented 2nd
arabic = TonedScale(tonic="D4", system=SYSTEMS["arabic"])
hijaz = arabic["hijaz"]
print(f"Maqam Hijaz from D: {hijaz.note_names}")

# Japanese hirajoshi -- hauntingly beautiful pentatonic
japanese = TonedScale(tonic="A4", system=SYSTEMS["japanese"])
hirajoshi = japanese["hirajoshi"]
print(f"Hirajoshi from A:   {hirajoshi.note_names}")
```

```python
# Blues scales -- the foundation of rock, jazz, and R&B
blues = TonedScale(tonic="A4", system=SYSTEMS["blues"])
print("Blues scales from A:")
for name in blues.scales:
    scale = blues[name]
    print(f"  {name:20s} -> {scale.note_names}")

# Gamelan -- approximations of non-Western tuning
gamelan = TonedScale(tonic="C4", system=SYSTEMS["gamelan"])
slendro = gamelan["slendro"]
print(f"\nGamelan slendro from C: {slendro.note_names}")
```

## 8. Guitar and Instruments

The `Fretboard` class models stringed instruments. You can look up
chord fingerings, render tab diagrams, apply a capo, and visualize
scale patterns across the neck.

```python
# Standard guitar fretboard
guitar = Fretboard.guitar()
print(f"Standard tuning: {guitar}")

# Look up chord fingerings from the chart
c_chart = CHARTS["western"]["C"]
print(f"\n{c_chart.tab(fretboard=guitar)}")
```

```python
# Show several common chord shapes
for chord_name in ["G", "Am", "Em", "D"]:
    chart = CHARTS["western"][chord_name]
    print(chart.tab(fretboard=guitar))
    print()
```

```python
# Apply a capo -- raises all strings by N semitones
capo2 = Fretboard.guitar(capo=2)
print(f"Capo on fret 2: {capo2}")
print("Playing 'G shape' with capo 2 = A major voicing")
```

```python
# Scale diagram -- see where notes fall on the neck
c_major_scale = TonedScale(tonic="C4")["major"]
diagram = guitar.scale_diagram(c_major_scale, frets=12)
print("C major scale on guitar:")
print(diagram)
```

## 9. Building a Song

Let's put it all together: pick a key, explore its chords, build a
progression, and analyze the harmonic movement.

```python
# Step 1: Choose a key
song_key = Key("E", "minor")
print(f"Key: {song_key}")
print(f"Notes: {song_key.note_names}")
print(f"Relative major: {song_key.relative}")
print(f"Signature: {song_key.signature}")
```

```python
# Step 2: See what chords are available
print("Diatonic chords in E minor:")
for i, name in enumerate(song_key.chords, 1):
    print(f"  {i}. {name}")

print("\nBorrowed chords from E major:")
for name in song_key.borrowed_chords[:4]:
    print(f"  {name}")
```

```python
# Step 3: Build a progression
# i - VI - III - VII is a classic minor key progression
prog = song_key.progression("i", "VI", "III", "VII")

print("Progression: i - VI - III - VII")
for chord in prog:
    name = chord.identify()
    numeral = chord.analyze("E", "minor")
    t = chord.tension
    print(f"  {name:18s}  [{numeral:5s}]  tension={t['score']:.2f}")
```

```python
# Step 4: Analyze voice leading through the progression
print("Voice leading through the progression:\n")
for i in range(len(prog) - 1):
    src = prog[i]
    dst = prog[i + 1]
    vl = src.voice_leading(dst)
    total = sum(abs(m) for _, _, m in vl)
    print(f"{src.identify()} -> {dst.identify()} (total movement: {total} semitones)")
    for s, d, m in vl:
        print(f"    {s} -> {d} ({m:+d})")
    print()
```

```python
# Step 5: Show the chords on guitar
guitar = Fretboard.guitar()
chord_names = ["Em", "C", "G", "D"]

print("Guitar charts for the progression:\n")
for name in chord_names:
    chart = CHARTS["western"][name]
    print(chart.tab(fretboard=guitar))
    print()
```

```python
# Bonus: Detect the key from a set of notes
detected = Key.detect("E", "G", "A", "B", "D")
print(f"Key detected from [E, G, A, B, D]: {detected}")

# Secondary dominant -- adds harmonic color
v_of_v = song_key.secondary_dominant(5)
print(f"\nSecondary dominant V/V in E minor: {v_of_v.identify()}")
print(f"Tension score: {v_of_v.tension['score']:.2f}")
```
